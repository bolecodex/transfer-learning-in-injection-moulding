```python
for i in df['Name']:
    Address='./3x2_Lego_%s_ZZP' % i #The address of loading the datasets
    Address2='./DataFrame_with_5_quality_value/3x2_Lego_%s_ZZP.csv' % i #The address of saving the datasets
    os.chdir(Address)
    
    df = pd.read_csv("DOE2_001.txt",engine='python') #Load the data
    df.rename(columns={'3x2 Lego':'id' }, inplace=True)# Rename the column
    print (df)
    a=[]
    for row in df.itertuples(): #Traverse all the rows
        #print(getattr(row, 'id'))
        row_split=getattr(row, 'id').split('  ')# Split the data by space
        #print(row_split)
        row_split_withoutnull = [i for i in row_split if i != ''] #Delete the empty element
        #print(row_split_withoutnull)
        a.append(row_split_withoutnull) #Make a list and add each row to the list
        #print(a)
    df=pd.DataFrame(a) #transfer the list to a dataframe
    #df.columns = df.iloc[0]# Rename the column
    
    df_quality_example=pd.DataFrame(df[2].tail(6)).T#Select the last 6 rows of current_value and do the transpose
    df_quality_example.columns = ['Length','Angle','Roundness','Height','Width','Part weight']#Define the column name
    df_quality_example.index = ['DOE2_001']#Define the index name
    df_quality_example
    
    filelist=[]# Initialization
    for root, dirs, files in os.walk(".", topdown=False): #find all the txt file in the folder
        for name in files:
            str=os.path.join(root, name)
            if str.split('.')[-1]=='txt':
                filelist.append(str)
    filelist
    
    pd.set_option('display.max_rows', None) #Show all rows in pandas
    pd.set_option('display.max_columns', None) # Show all columns in pandas
    file_chdir = os.getcwd()# Initialization of the folder address
    file_txt = []
    for root, dirs,files in os.walk(file_chdir): #Also find all the excel file in the folder, Alternative
        for file in files:
            if os.path.splitext(file)[1] =='.txt':
                file_txt.append(file)
    df_quality=df_quality_example.drop(index=['DOE2_001'])# Initialization
    
    
    pd.set_option('display.max_rows', None) #Show all rows in pandas
    pd.set_option('display.max_columns', None) # Show all columns in pandas
    file_chdir = os.getcwd()# Initialization of the folder address
    file_txt = []
    for root, dirs,files in os.walk(file_chdir): #Also find all the excel file in the folder, Alternative
        for file in files:
            if os.path.splitext(file)[1] =='.txt':
                file_txt.append(file)
    df_quality=df_quality_example.drop(index=['DOE2_001'])# Initialization
    Subname='DOE2'
    for file in file_txt:
        result = Subname in file #Select the "DOE2_0xx.txt" data
        if result == True:
            df = pd.read_csv(file,engine='python') #Load the data
            df.rename(columns={'3x2 Lego':'id' }, inplace=True)# Rename the column
            a=[] # Initialization
            for row in df.itertuples(): #Traverse all the rows
                row_split=getattr(row, 'id').split('  ')# Split the data by space
                row_split_withoutnull = [i for i in row_split if i != ''] #Delete the empty element
                a.append(row_split_withoutnull) #Make a list and add each row to the list
            df=pd.DataFrame(a) #transfer the list to a dataframe
            df_quality_example=pd.DataFrame(df[2].tail(6)).T#Select the last 6 rows of current_value and do the transpose
            df_quality_example.index = [file.strip('.txt')] #Define the index name
            df_quality_example.columns = ['Length','Angle','Roundness','Height','Width','Part weight']
            df_quality=pd.concat([df_quality, df_quality_example]) # Concate all the data 
    df_quality

    lines = open("DOE1.txt","rb").readlines() #Load the file
    lines_Machine=lines[5:13] #Find the lines which represent machine parameters
    df=pd.DataFrame(lines_Machine) #Transfer the list to dataframe
    print(df)
    a=[]
    for row in df.iterrows(): #Traverse all the rows
        row_split=row.__str__().split('  ')  #Transfer the object to the str and split the str by space
        row_split_withoutnull = [i for i in row_split if i != ''] #Delete the empty element
        #print (row_split_withoutnull)
        a.append(row_split_withoutnull) #Make a list and add each row to the list
    df=pd.DataFrame(a) #Make a dataframe
    
    df_set_machine_value=pd.DataFrame(df[[4,5]].tail(6))#Select the last 6 rows of current_value,which are the machine parameters
    df_set_machine_value.index = range(6)#Reset the index
    df_set_machine_value.rename(columns={4:'value' ,5:'step_value'}, inplace=True) #Rename the column
    df_set_machine_name=pd.DataFrame({'Machine parameters':['Melt temperature','Cavity wall temperature','Holding pressure time','Cooling time','Injection flow rate','Holding pressure']})
    df_set_machine= pd.concat([df_set_machine_name,df_set_machine_value], axis=1).T #Concate the two data set
    df_set_machine.columns = df_set_machine.iloc[0]# Rename the column as the first row
    df_set_machine=df_set_machine.drop(['Machine parameters']) #Delete the useless index column
    df_set_machine=df_set_machine.astype(float) #Change the object format to float
    df_set_machine.loc['+'] = df_set_machine.apply(lambda x: x.loc['value'] + x.loc['step_value']) #Add a row "+" which is caluculated by the summation of the first two rows
    df_set_machine.loc['-'] = df_set_machine.apply(lambda x: x.loc['value'] - x.loc['step_value']) #Add a row "-" which is caluculated by the difference of the first two rows
    df_set_machine
    
    os.chdir(os.path.dirname(os.getcwd())) # Go back two times to find the file address
    df=pd.read_csv('./Versuchsplan1.txt',engine='python')#Experiment plan1 dataset
    df_experimentplan1=df['Nr'].str.split(' ',expand=True) #Split the file by space
    df_experimentplan1.columns = ['Melt temperature','Cavity wall temperature','Holding pressure time','Cooling time','Injection flow rate','Holding pressure']
    df_experimentplan1
    
    df_machine=pd.concat([df_experimentplan1, df_set_machine]) #Combine the experiment plan1 and machine parameters
    for i in range(len(df_machine)-4): #Traverse all the rows of experiment plan1
        for j in df_machine.columns: #Tanverse all the columns
            if df_machine[j].loc[i]=='+': #If it is the experiment plan is "+", We replace it by the corresponding number in the '+' row
                df_machine[j].loc[i]=df_machine[j].loc['+']
            else: #If it is the experiment plan is "-", We replace it by the corresponding number in the '-' row
                df_machine[j].loc[i]=df_machine[j].loc['-']
    df_machine.loc[64]=df_machine.loc['value'] # Because the values in experiment plan are all 0, It shoube the set value.
    df_machine
    
    df_quality.reset_index(drop=True, inplace=True)
    Example_result = pd.concat([df_quality[['Part weight','Length', 'Angle','Height','Width']], df_machine.iloc[:65] ],axis=1)
    Example_result.to_csv(Address2)
    
    
    #os.chdir(os.path.dirname(os.getcwd()))#Return to the parent folder
    print (Address)
```
